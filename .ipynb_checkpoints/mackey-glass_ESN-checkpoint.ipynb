```python
import reservoirpy.nodes as nodes
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import reservoirpy as rpy
from reservoirpy.datasets import mackey_glass
rpy.set_seed(42)  
```

```python
X = mackey_glass(2000, tau=17)

# Normalizing Series in the range of [-1,1]
X_min, X_max = X.min(), X.max()
X_norm_01 = (X - X_min) / (X_max - X_min)
X_norm_final = 2 * X_norm_01 - 1
```

```python
X_norm_final.max()
```

```python
plt.figure(figsize=(14,4))
plt.plot(X_norm_final)
plt.show()
```

```python
reservoir = nodes.Reservoir(500, sr=1, lr=0.1, seed=42)
readout = nodes.Ridge(ridge=1e-4)

esn_model = reservoir >> readout
```

```python
X_train = X_norm_final[:1600]
Y_train = X_norm_final[1:1601]

esn_model = esn_model.fit(X_train, Y_train, warmup=50)
```

```python
Y_pred_5 = np.empty((5, 1))
x = Y_train[-1].copy().reshape(1,1)
STATE_pred_5 = np.empty((5, 500))
for i in range(5):
    # print(x)
    current_state = reservoir.run(x)
    STATE_pred_5[i, :] = current_state
    
    Y_pred_5[i] = readout(current_state.squeeze())
    x = Y_pred_5[i].reshape(1,1)
# Y_pred_5
```

```python
# _ = esn_model.reset()
```

```python
fig, axes = plt.subplots(2,1, figsize=(10, 8))

axes[0].set_title("Mackey-Glass")
axes[0].set_xlabel("$t$")
axes[0].plot(Y_pred_5, label="Predicted (t+1)", color="blue")
axes[0].plot(X_norm_final[1600:1606], label="Real (t+1)", color="red")

axes[1].set_title("Model State")
axes[1].set_xlabel("$t$")
axes[1].plot(STATE_pred_5[:, :20])

plt.legend()
plt.show()
```
