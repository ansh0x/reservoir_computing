```python
import reservoirpy.nodes as nodes
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import reservoirpy as rpy
rpy.set_seed(42)  
```

```python
reservoir = nodes.Reservoir(100, lr=0.5, sr=0.9)
```

```python
X = np.sin(np.linspace(0, 6*np.pi, 100)).reshape(-1,1)
```

```python
plt.figure(figsize=(10, 3))
plt.title("Sine Wave")
plt.ylabel("$sin(t)$")
plt.xlabel("$t$")
plt.plot(X)
plt.show()
```

```python
s = reservoir(X[0])
```

```python
states = reservoir.run(X)
```

```python
_ = reservoir.reset()
```

```python
readout = nodes.Ridge(ridge=1e-7)
```

```python
X_train = X[:50]
Y_train = X[1:51]
```

```python
train_state = reservoir.run(X_train)
```

```python
train_state.shape
```

```python
readout = readout.fit(train_state, Y_train, warmup=10)
```

```python
test_states = reservoir.run(X[50:])
Y_pred = readout.run(test_states)

plt.figure(figsize=(10, 3))
plt.title("Sine Wave prediction")
plt.xlabel("$t$")
plt.ylabel("$sine(t)$")
plt.plot(Y_pred, label="Predicted sin(t)", color="blue")
plt.plot(X[51:], label="Real sin(t+1)", color="red")
plt.legend()
plt.show()
```

# Train the ESN

```python
reservoir = nodes.Reservoir(100, lr=0.5, sr=0.99)
readout = nodes.Ridge(ridge=1e-2)

esn_model = reservoir >> readout
```

```python
esn_model = esn_model.fit(X_train, Y_train, warmup=10)
```

```python
Y_pred = esn_model.run(X[50:])
print(Y_pred[:5])
plt.figure(figsize=(10, 3))
plt.title("A sine wave and its future.")
plt.xlabel("$t$")
plt.plot(Y_pred, label="Predicted sin(t+1)", color="blue")
plt.plot(X[51:], label="Real sin(t+1)", color="red")
plt.legend()
plt.show()
```

```python
Y_pred = np.empty((100, 1))
x = Y_train[-1]

for i in range(100):
    x = esn_model(x)
    Y_pred[i] = x
```

```python
plt.figure(figsize=(10, 3))
plt.title("A sine wave and its future.")
plt.xlabel("$t$")
plt.plot(Y_pred, label="Predicted sin(t+1)", color="blue")
plt.plot(X, label="Real sin(t+1)", color="red")
plt.legend()
plt.show()
```
